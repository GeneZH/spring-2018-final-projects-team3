```python
import csv
filename = "./stop_times.txt"
trips_stops = {}
with open(filename) as f:
    reader = csv.reader(f)
    data_list = list(reader)
    for row in data_list[1:]: 
        if row[0] not in trips_stops:
            trips_stops[row[0]] = [row[3]]
        else:
            trips_stops[row[0]].append(row[3])
print(len(trips_stops))     
```

```python
import csv
filename = "./trips.txt"
routes_trips = {}
with open(filename) as f:
    reader = csv.reader(f)
    data_list = list(reader)
    for row in data_list[1:]: 
            routes_trips[row[1]] = row[-1]
print(len(routes_trips)) 
print(routes_trips)
```

```python
routes_stops = {}
for r,t in routes_trips.items():
    routes_stops[r] = trips_stops[t]
print(routes_stops)
    
```

```python
import csv
filename = "./stops.txt"
stops_locations = {}
with open(filename) as f:
    reader = csv.reader(f)
    data_list = list(reader)
    for row in data_list[1:]: 
        stops_locations[row[-2]] = [row[0], row[3]]
print(stops_locations)    
```

```python
import csv
csvFile2 = open('routes.csv','w', newline='')
writer2 = csv.writer(csvFile2)
for r,s in routes_stops.items():
    writer2.writerow(r)
    latlng = []
    for stop in s:
        latlng.append(stops_locations[stop])
    csvFile = open('./stops/' + r + '.csv','w', newline='')
    writer = csv.writer(csvFile)
    for location in latlng:
        writer.writerow(location)
    csvFile.close()
csvFile2.close()
        


```
